2) Change data type

```python
def convert_dtypes(df):

    # Convert air quality and weather columns to float
    float_cols = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP", "RAIN", "WSPM"]
    for col in float_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")  # Convert to float, handle errors

    return df

# Apply the function to df_combined
df_combined = convert_dtypes(df_combined)

# Verify data types after conversion
print(df_combined.dtypes)
```

3) Add timestamp column having datetime format **(ISO 8601)**

```python
df_combined['timestamp'] = pd.to_datetime(df_combined[['year', 'month', 'day', 'hour']])
```

4-a) Fill missing values

```python
def fill_missing_values(df):

    df_filled = df.copy()

    # Fill numeric columns with the mean
    df_filled.update(df_filled.select_dtypes(include=['number']).fillna(df_filled.mean(numeric_only=True)))

    # Fill categorical/string columns with the mode (most frequent value)
    for col in df_filled.select_dtypes(include=['object']).columns:
        mode_value = df_filled[col].mode()[0] if not df_filled[col].mode().empty else ""
        df_filled[col].fillna(mode_value, inplace=True)

    return df_filled

# Apply the function
df_combined = fill_missing_values(df_combined)
# Display the cleaned DataFrame
#print("Final DataFrame shape:", df_combined.shape)
#df_combined.head(3)
```

```python
df_combined.isnull().sum()
```

5) Normalize  or 	Standardize Measurement Units

```python
# Define the Z-score normalization function
def z_score_normalization(df, columns):
    for column in columns:
        mean = df[column].mean()
        std = df[column].std()
        df[column] = (df[column] - mean) / std
    return df

# List of pollutant columns to normalize
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

# Apply Z-score normalization
df_combined = z_score_normalization(df_combined, pollutants)
```

```python
# Optional: View summary statistics to confirm normalization
df_combined[pollutants].describe()
```

6) Data aggregation: Add column in the final dataframe Based on a Condition (e.g., Air Quality Category)

```python
def classify_air_quality(pm25):
    if pm25 <= 12:
        return 'Good'
    elif pm25 <= 35.4:
        return 'Moderate'
    elif pm25 <= 55.4:
        return 'Unhealthy for Sensitive Groups'
    elif pm25 <= 150.4:
        return 'Unhealthy'
    else:
        return 'Very Unhealthy'

df_combined['PM2.5_category'] = df_combined['PM2.5'].apply(classify_air_quality)
```
